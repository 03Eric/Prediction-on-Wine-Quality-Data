# Checking the Wine Quality distribution

```python
def feature_viz(feature):
    plt.figure(figsize=(15,8))
    plt.title(f'{feature} hist plot')
    plt.subplot(1,3,1)
    wine_df[feature].plot(kind='hist')

    plt.subplot(1,3,2)
    plt.title(f'{feature} box plot')
    sns.boxplot(wine_df[feature])
    
    plt.subplot(1, 3, 3)
    plt.title(f'{feature} density plot')
    sns.kdeplot(wine_df[feature])
    plt.tight_layout()
    
```

# Checking whether the distribution is normal and checking the number of outliers

```python
for i in wine_df.columns:
    feature_viz(i)
```

As above we can see there are extreme outliers in some features.

# Checking correlation through heat map 

```python
wine_df_corr = wine_df.corr(method="spearman")
plt.figure(figsize=(22,22))
sns.heatmap(wine_df_corr, vmin=-1, vmax=1, cmap="Blues", annot=True, linewidth=0.1)
```

```python
# taking Main Features having correlation more than 0.20 and less than -0.40
main_features=['citric acid','sulphates','alcohol','quality','volatile acidity']
```

# Model Building

```python
#Creating X and Y.
X= wine_df[main_features].copy()
X=X.drop(['quality'], axis=1)
y= wine_df.quality
```

```python
#Splitting the data into train and test.
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
print(len(X_train),len(y_train),len(X_test),len(y_test))
```

# Predicting Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model_1=LogisticRegression()
model_1.fit(X_train,y_train)
predictions_1=model_1.predict(X_test)
print(accuracy_score(y_test, predictions_1))
```

# Predicting Using RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
model_2 = RandomForestClassifier()
model_2.fit(X_train, y_train)
predictions_2=model_2.predict(X_test)
print(accuracy_score(y_test, predictions_2))
```

# Predicting Using DecisionTreeClassifier

```python
from sklearn import tree
model_3=tree.DecisionTreeClassifier()
model_3.fit(X_train,y_train)
predictions_3=model_3.predict(X_test)
print(accuracy_score(y_test, predictions_3))
```
